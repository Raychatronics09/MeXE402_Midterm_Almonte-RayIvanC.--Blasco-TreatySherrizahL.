## Linear Regression

### Part 1 - Data Preprocessing

### Importing the dataset

```python
import pandas as pd

dataset = pd.read_csv('layfraw.csv')

print(dataset.columns)

#variable and this is a function for uploading the dataset
```

```python
#display 5 rows of dataset

#10,000 rows
#data points collected from a combined cycle power plant over six years
#5 columns: AT ambient temp,V exhaust vacuum, AP ambient pressure, RH relative humdity, PE net hourly  electrical energy output
# independent variables: AT, V, AP and RH
# dependent variable: PE

# Fill missing values in numerical columns with the column mean

dataset['Adult Mortality'].fillna(dataset['Adult Mortality'].mean(), inplace=True)
dataset['Alcohol'].fillna(dataset['Alcohol'].mean(), inplace=True)
dataset['Hepatitis B'].fillna(dataset['Hepatitis B'].mean(), inplace=True)
dataset['BMI'].fillna(dataset['BMI'].mean(), inplace=True)
dataset['Polio'].fillna(dataset['Polio'].mean(), inplace=True)
dataset['Total expenditure'].fillna(dataset['Total expenditure'].mean(), inplace=True)
dataset['Diphtheria'].fillna(dataset['Diphtheria'].mean(), inplace=True)
dataset['GDP'].fillna(dataset['GDP'].mean(), inplace=True)
dataset['Population'].fillna(dataset['Population'].mean(), inplace=True)
dataset['thinness 10-19 years'].fillna(dataset['thinness 10-19 years'].mean(), inplace=True)
dataset['thinness 5-9 years'].fillna(dataset['thinness 5-9 years'].mean(), inplace=True)
dataset['Income composition of resources'].fillna(dataset['Income composition of resources'].mean(), inplace=True)
dataset['Schooling'].fillna(dataset['Schooling'].mean(), inplace=True)
dataset['Life expectancy'].fillna(dataset['Life expectancy'].mean(), inplace=True)

dataset.head(10)
```

### Getting the inputs and output

```python
# [rows,columns]
X = dataset.iloc[:,3:-1].values
X
```

```python
y = dataset.iloc[:,-1].values
y
```

### Creating the Training Set and the Test Set

```python
# scikitlearn is a library
# model_selection is a module
# train_test_split is a function

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=0)
```

## Part 2 - Building and training the model

### Building the model

```python
# linear_model is the module
# `LinearRegression is a class` is defining that `LinearRegression` is a class within the `linear_model` module. It indicates that `LinearRegression` is a blueprint or template for creating objects that represent linear regression models.
# Class is a pre-coded blueprint of something we want to build from which objects are created.

from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

### Training the Model

```python
# fit is a method inside LinearRegression class - they are like functions.

model.fit(X_train,y_train)
```

### Inference

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test
```

#### Making the prediction of a single data point with AT = 15, V = 40, AP = 1000, RH = 75

```python

PE = model.predict([[214,54,4.52,31.27232188,67,58848,24.8,68,77,4.71,67,0.1,2842.938353,112249,1,9.7,0.676,11.7]])
PE
```

## Part 3: Evaluating the Model

### R-Squared

```python


# Function = small letter
# Class = Big Letter

from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)

r2
# Very Good
# Should be very good for most times
# if not, then maybe a different method of analysis should be used or error in code
```

### Adjusted R-Squared

```python

k = X_test.shape[1]
k
```

```python
n = X_test.shape[0]
n
```

```python

adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)
adj_r2
```
